# Find roots and local extrema with `scipy` 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a dummy graph
x1 = np.linspace(1, 10, 100)
x2 = np.linspace(10, 15, 100)
x3 = np.linspace(15, 20, 100)
y1 = 2 * np.sin(x1)
y2 = 3 * np.sin(-1.54 * x2)
y3 = 1 * np.sin(2.5 * x3)
y = np.append(np.append(y1, y2), y3)
x = np.append(np.append(x1, x2), x3)

# Create y-coordinates for line y = 0
y0 = x * 0.0

plt.plot(x, y, '-', color='k')
plt.plot(x, y0, '--', color='k', lw=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Random Graph')
plt.show()
```

```python
from scipy.signal import find_peaks  
import numpy as np
import matplotlib.pyplot as plt      

# NOTE: 
# Lists 'x' ,'y' of the curve are defined above
# List 'y0' representing y=0 also defined above

# Find index of each extrema
local_max, _ = find_peaks(y)    # Local max points
local_min, _ = find_peaks(-y)   # Local min points

# Find indices of intersections between curve and y=0
idx = np.argwhere(np.diff(np.sign(y0 - y))).flatten()

# Create plot of graph with local extrema and roots
plt.plot(x, y, '-', color='k')                                                # Plot curve
plt.plot(x[local_max], y[local_max], 'rx', markersize=10, label='Local max')  # Plot local max
plt.plot(x[local_min], y[local_min], 'bx', markersize=10, label='Local min')  # Plot local min
plt.plot(x[idx], y0[idx], 'o', label='Roots', color='magenta')                # Plot roots
plt.legend()                                                                  # Add legend
plt.show()                                                                    # Show graph  
```
